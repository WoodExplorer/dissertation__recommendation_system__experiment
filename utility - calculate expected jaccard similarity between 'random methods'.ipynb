```python
import bigfloat as bf
from operator import mul    # or mul=lambda x,y:x*y
from fractions import Fraction
```

```python

def nCk(n,k): 
    return int( reduce(mul, (Fraction(n-i, i+1) for i in range(k)), 1) )
```

```python
size_of_chosen = 400
total = 4000
size_of_not_chosen = total - size_of_chosen

for x in range(1, size_of_chosen + 1):
    x, size_of_chosen, size_of_chosen - x, size_of_not_chosen, '\t', x, size_of_chosen + size_of_chosen - x
    print x, size_of_chosen, size_of_chosen - x, size_of_not_chosen, '\t', x, size_of_chosen + size_of_chosen - x
```

```python
def calculate_expected_jaccard(total, size_of_chosen):
    acc = bf.BigFloat(0)
    size_of_not_chosen = total - size_of_chosen

    for step, x in enumerate(range(1, size_of_chosen + 1)):
        if (0 == step % 100):
            print 'progress:', step

        a, b, c, d, e, f = x, size_of_chosen, size_of_chosen - x, size_of_not_chosen, x, size_of_chosen + size_of_chosen - x
        acc += bf.BigFloat(nCk(b, a)) * bf.BigFloat(nCk(d, c)) * e / f

    print 'accumulation:', acc
    
    expected_jaccard_if_random = acc / nCk(total, size_of_chosen)
    print 'expected_jaccard_if_random:', expected_jaccard_if_random
    return expected_jaccard_if_random
```
