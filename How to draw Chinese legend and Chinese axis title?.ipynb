```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

df = pd.DataFrame( data=[[1,2], [10, 20]], columns=[u'é',u'日本'] )

ax = df.plot()
legend = ax.legend()
font = font_manager.FontProperties(fname='/home/wsyj/Downloads/simsun.ttc')

for text in legend.texts:
    text.set_font_properties(font)
    
plt.xlabel(u'中文', fontproperties=font)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

df = pd.DataFrame( data=[[1,2], [10, 20]], columns=[u'é',u'日本'] )

ax = df.plot()
legend = ax.legend()
font = font_manager.FontProperties(fname='/home/wsyj/Downloads/simsun.ttc')

for text in legend.texts:
    text.set_font_properties(font)

ax.set_xlabel(u'中')

for label in ax.get_xticklabels() :
    label.set_fontproperties(font)
    
plt.xlabel(u'中文', fontproperties=font)
#label = ax.get_xlabel()
#print label
#label = ax.get_label()
#print label
#print label.set_font_properties(font)
#
plt.show()
```
