```python
# With help of subplot, we can apply different markers to points on different lines.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

fig1 = plt.figure()

df = pd.DataFrame( data=[1,2], columns=[u'é'] ) # It seems to me that when using underneath pyplot directly, the columns here will not translate to legend.

ax1 = fig1.add_subplot(111)
ax1.plot(df, label=u'é', marker='o')

df = pd.DataFrame( data=[10, 20], columns=[u'日本'] )

ax2 = fig1.add_subplot(111)
ax2.plot(df, label=u'日本', marker='*')


legend = plt.legend()
font = font_manager.FontProperties(fname='/home/wsyj/Downloads/simsun.ttc')

for text in legend.texts:
    text.set_font_properties(font)

#plt.set_xlabel(u'中')
#
#for label in ax.get_xticklabels() :
#    label.set_fontproperties(font)
    
plt.xlabel(u'中文', fontproperties=font)

plt.show()

```
